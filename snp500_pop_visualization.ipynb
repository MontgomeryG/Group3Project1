```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading In Snp500
df = pd.read_csv('../Group3Project1/Resources/snp500_no_nan_population.csv')

# How many companies per GICS Sector
sector_counts = df['GICS Sector'].value_counts()
print(sector_counts)
```

```python
# Function to create a bar chart for a given sector
def create_sector_bar_chart(data, sector):
  sector_data = data[data['GICS Sector'] == sector]
  sector_data = sector_data.sort_values(by='population', ascending=False)
  plt.figure(figsize=(10, 6))
  bars = plt.bar(sector_data['Security'], sector_data['population'], color='skyblue')
  # Creating City Labels for the Bars in chart
  # Loop over each bar and the corresponding row in the DataFrame
  for bar, (_, row)in zip(bars, sector_data.iterrows()):
    # Get the height of the current bar
    y_value = bar.get_height()
    # Adding text onto each bar with the city name
    plt.text(bar.get_x() + bar.get_width()/2, y_value*.80, row['city'],
             rotation=90, verticalalignment='bottom', fontsize=7, ha='center') 
  plt.xlabel('Company')
  plt.ylabel('City Population')
  plt.title(f'City Population of Companies in {sector} Sector', y=1.1)
  plt.xticks(rotation=90)
  plt.yscale('log')
  plt.tight_layout()
  plt.show()
```

```python
# For Loop for each sector
for sector in df['GICS Sector'].unique():
  create_sector_bar_chart(df, sector)
```
