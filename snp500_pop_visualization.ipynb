```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading In Snp500
df = pd.read_csv('../Group3Project1/Resources/snp500_no_nan_population.csv')

# How many companies per GICS Sector
sector_counts = df['GICS Sector'].value_counts()
print(sector_counts)
```

```python
# Function to create a bar chart for a given sector
def create_sector_bar_chart(data, sector):
  sector_data = data[data['GICS Sector'] == sector]
  sector_data = sector_data.sort_values(by='population', ascending=False)
  plt.figure(figsize=(10, 6))
  plt.bar(sector_data['Security'], sector_data['population'], color='skyblue')
  plt.xlabel('Company')
  plt.ylabel('City Population')
  plt.title(f'City Population of Companies in {sector} Sector')
  plt.xticks(rotation=90)
  plt.yscale('log')
  plt.tight_layout()
  plt.show()
```

```python
# For Loop for each sector
for sector in df['GICS Sector'].unique():
    create_sector_bar_chart(df, sector)
```
