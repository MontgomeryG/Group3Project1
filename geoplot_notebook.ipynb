Now, let's bring in our S&P 500 average yearly return data for 2022 and add that to our new dataframe containing locations of the companies.

```python
# Read in .csv containing 2022 average daily returns.
average_returns_df = pd.read_csv(Path('../Group3Project1/Resources/average_daily_return_2022.csv'))

# Preview
average_returns_df.head()
```

```python
# Make symbol the index
average_returns_df = average_returns_df.rename(columns={'Unnamed: 0': 'symbol'})
average_returns_df.index = average_returns_df['symbol']
average_returns_df = average_returns_df.drop('symbol', axis=1)

average_returns_df.head()
```

```python
# Read in .csv containing 'Symbol' and 'Secruity'
tickers_with_names_df = pd.read_csv(Path('../Group3Project1/Resources/constituents.csv'))
tickers_with_names_df.head()
```

```python
# Capture only the symbol and security columns
symbol_security_df = pd.concat([tickers_with_names_df["Symbol"], tickers_with_names_df["Security"]], axis=1)


# Create symbol as index
symbol_security_df = symbol_security_df.rename(columns={'Symbol': 'symbol'})
symbol_security_df.index = symbol_security_df['symbol']
symbol_security_df = symbol_security_df.drop('symbol', axis=1)

symbol_security_df.head()
```

```python
# Sort alphabetically 
symbol_security_df.sort_index(inplace=True)
symbol_security_df.head()
```

```python
# Now we put these together
display(average_returns_df.head(2))
display(symbol_security_df.head(2))
```

```python
returns_security_df = pd.merge(average_returns_df, symbol_security_df, on='symbol', how='left')
returns_security_df
```

```python
# Now add the returns data onto our dataframe containing lat/lon
final_df = pd.merge(returns_security_df, merge_datasets, on=['Security'], how='left')
```

```python
final_df
```

With our current dataframe, now we can plot returns versus location. 

```python
# Getting an error saying the size cannot be negative. So we will use the absolute value function 
#   to make all return values positive. So the graph will instead only show us the magnitude of the returns
#   rather than the magnitude and direction of them.

final_df['2022 Average Daily Return'] = final_df['2022 Average Daily Return'].abs()
final_df
```

```python
# Only keep rows with the needed cities
final_df = final_df[final_df['city_x'] == final_df['city_y']]
len(final_df)
```

```python
# Only keep rows where these needed cities are in the correct states
final_df = final_df[final_df['state_x']==final_df['state_y']]
len(final_df)
```

```python
# Drop duplicates
final_drop_dup_df = final_df.drop_duplicates(subset='Security')
```

```python
final_drop_dup_df
```

NOW, let's try to plot.

```python





returns_location_plot = final_drop_dup_df.hvplot.points(
    'LONGITUDE', 
    'LATITUDE',
    geo=True, 
    title = 'Magnitude of Returns Given Location',
    #size="2022 Average Daily Return", 
    #size_max=25,
    color='2022 Average Daily Return',
    #zoom=11,
    #alpha=0.4, #transparency level of the bubbles
    tiles='OSM',        #OSM - Open Street Map
    frame_width = 700,
    frame_height = 500,
    xlabel='Longitude',
    ylabel='Latitude',
    )



returns_location_plot
```

The above shows the locations of each S&P 500 headquarters. The color of each bubble represents the magnitude of the return; the darker the blue the higher the magnitude of the return.

This is a good starting point, but in order to get more useful information we may start off by normalizing the return data so that more weight is not given to a given stock's return simply because it is a higher priced stock.
