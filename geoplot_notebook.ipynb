```python
# Import the required libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
import os
```

```python
# Using the read_csv function and Path module, create a DataFrame 
us_cities_df = pd.read_csv(
    Path('../Group3Project1/Resources/us_cities.csv')
)

# Review the first five rows of the DataFrame
display(us_cities_df.head())
display(us_cities_df.tail())
```

```python
# Using the read_csv function and Path module, create a DataFrame 
snp500_with_pop_df = pd.read_csv(
    Path('../Group3Project1/Resources/snp500_with_population.csv')
)

# Review the first five rows of the DataFrame
display(snp500_with_pop_df.head())
display(snp500_with_pop_df.tail())
```

```python
# create new column, first_word
# split method
# first name of city as index

snp500_with_pop_df['first_word'] = snp500_with_pop_df['city'].str.split().str[0]

display(snp500_with_pop_df.head())
display(snp500_with_pop_df.tail())
```

```python
us_cities_df['first_word'] = us_cities_df['CITY'].str.split().str[0]

display(us_cities_df.head())
display(us_cities_df.tail())
```

```python
us_cities_df.rename(columns={"STATE_NAME":"state"}, inplace=True)
```

```python
display(us_cities_df.head())
display(snp500_with_pop_df.head())
```

```python
# Merge datasets by State THEN by First Word of city

merge_datasets = pd.merge(snp500_with_pop_df, us_cities_df, on=['state', 'first_word'], how='left')
```
